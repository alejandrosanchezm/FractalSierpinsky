```python
def generar_fractal_sierpinski(iteraciones=10000):

    import random

    punto_a = [0,0]
    punto_b = [6,0]
    punto_c = [3,5]

    puntos_x = [0,6,3,3]
    puntos_y = [0,0,5,5/3]

    punto_actual = [3,5/3]

    def punto_random():
        val = random.randint(1,3)
        if val == 1:
            return punto_a
        if val == 2:
            return punto_b
        if val == 3:
            return punto_c

    def punto_medio(p1, p2):

        x1 = p1[0]
        y1 = p1[1]

        x2 = p2[0]
        y2 = p2[1]

        return [(x1+x2)/2, (y1+y2)/2]

    for i in range(iteraciones):

        pt = punto_random()
        result = punto_medio(pt, punto_actual)
        puntos_x.append(result[0])
        puntos_y.append(result[1])
        punto_actual = result
        
    return puntos_x, puntos_y
```

```python
import matplotlib.pyplot as plt

puntos_x, puntos_y = generar_fractal_sierpinski(100000)
plt.rcParams["figure.figsize"] = (10,10)

plt.scatter(puntos_x, puntos_y,c=['#8c564b'],s=0.1)
```
